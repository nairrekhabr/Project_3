```python
# Import libraries and dependencies
import os
import pandas as pd
import numpy as np
from MCForecastTools import MCSimulation
from pathlib import Path
```

```python
# Load csv with LTC daily closing prices and returns and format into dataframe compatible with MCForecastTools.py
ltc_df = pd.read_csv(Path("ltc_daily_returns.csv"), index_col = [0], header = [0,1])
ltc_df = ltc_df.dropna()
ltc_df

#Confirm that there are no null values in the dataframe
#print(ltc_df.isnull().sum())
```

```python
# Load csv with MKR daily closing prices and returns and format into dataframe compatible with MCForecastTools.py
mkr_df = pd.read_csv(Path("mkr_daily_returns.csv"), index_col = [0], header = [0,1])
mkr_df = mkr_df.dropna()
mkr_df


#Confirm that there are no null values in the dataframe
#print(mkr_df.isnull().sum())
```

```python
# Set up 1000-path, 10-year Monte Carlo simulations for LTC & MKR
MC_ltc = MCSimulation(
    portfolio_data = ltc_df,
    weights = [1],
    num_simulation = 1000,
    num_trading_days = 252*10
)

MC_mkr = MCSimulation(
    portfolio_data = mkr_df,
    weights = [1],
    num_simulation = 1000,
    num_trading_days = 252*10
)
```

```python
#Running a monte carlo simulation to forecast 10=year cumulative returns

MC_ltc.calc_cumulative_return()
MC_mkr.calc_cumulative_return()
```

```python
# Fetch summary statistics from the Monte Carlo simulation results
ltc_stats = MC_ltc.summarize_cumulative_return()
mkr_stats = MC_mkr.summarize_cumulative_return()

# Print summary statistics
print(f"Litecoin Forecast Statistics:", '\n', ltc_stats, '\n')
print(f"Maker Forecast Statistics:", '\n', mkr_stats, '\n')
```

```python
# Define initial investment amount
initial_investment = 10000

# Define LTC confidence intervals & median
ltc_ci_lower = round(ltc_stats[8] * initial_investment, 2)
ltc_ci_upper = round(ltc_stats[9] * initial_investment, 2)
ltc_median = round(ltc_stats[5] * initial_investment, 2)

# Define MKR confidence intervals & median
mkr_ci_lower = round(mkr_stats[8] * initial_investment, 2)
mkr_ci_upper = round(mkr_stats[9] * initial_investment, 2)
mkr_median = round(mkr_stats[5] * initial_investment, 2)

# Print confidence interval and median summaries for all entities in comparison
print(f"LTC Forecast Summary:")
print(f"There is a 95% chance that an initial investment of ${initial_investment} in LTC"
      f" will be worth between ${ltc_ci_lower} and ${ltc_ci_upper} in 10 years.")
print(f"The median expected value of a current ${initial_investment} investment in LTC "
      f"is ${ltc_median} in 10 years.", '\n')

print(f"MAKER Forecast Summary:")
print(f"There is a 95% chance that an initial investment of ${initial_investment} in Maker"
      f" will be worth between ${mkr_ci_lower} and ${mkr_ci_upper} in 10 years.")
print(f"The median expected value of a current ${initial_investment} investment in Maker "
      f"is ${mkr_median} in 10 years.", '\n')
```
