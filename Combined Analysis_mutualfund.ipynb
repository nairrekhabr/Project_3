```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
import numpy as np
import yfinance as yf
from yahoofinancials import YahooFinancials
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
import hvplot.pandas
```

```python
## Set the SP500 tickers for which data has to be obtained
assets_sp500 = ['PREIX', 'VOO', 'FXAIX','SWPPX']

# Call the yahoo financial API 
yahoo_financials = YahooFinancials(assets_sp500)

# Call the get historical price data function between the stipulated time range
data = yahoo_financials.get_historical_price_data(start_date='2017-11-01', 
                                                  end_date='2021-11-02', 
                                                  time_interval='Daily')

#Build the dataframe from thee obtained historical data
prices_df_sp500 = pd.DataFrame({
    a: {x['formatted_date']: x['adjclose'] for x in data[a]['prices']} for a in assets_sp500
})

# Set the legend label values and plot the multiple value graph
ax_sp500=prices_df_sp500.plot(y=['PREIX','VOO','FXAIX','SWPPX'],label=["PREIX[T.ROWE]","VOO[Vanguard]","FXAIX[FIDELITY]","SWPRX[Schwab]"],title="Comparative Analysis SP500")

# Set the x & Y axis labels
ax_sp500.set(xlabel='Date', ylabel='Adj.close Prices')
```
